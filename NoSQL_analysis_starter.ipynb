### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq':20}}

# Use count_documents to display the number of documents in the result
hyg_result = list(estab.find(query))
hyg20 = estab.count_documents(query)
print(f'There are {hyg20} business with a hygiene score of 20.')
# Display the first document in the results using pprint
pprint(hyg_result[0])
```

```python
# Convert the result to a Pandas DataFrame
hyg20_df = pd.DataFrame(hyg_result)

# Display the number of rows in the DataFrame
print(len(hyg20_df))

# Display the first 10 rows of the DataFrame
hyg20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
estab.update_many({}, [ {'$set':{ 'RatingValue' : {'$toInt': "$RatingValue"}}} ])
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

match_query = {
                'LocalAuthorityName': {'$regex':'London'},
                 'RatingValue':{'$gte':4}

                }



result = list(estab.find(match_query))
london4s = estab.count_documents(match_query)
print(f'There are {london4s} businesses in London with a rating greater than or equal to 4.')

pprint(result[0])

```

```python
# Convert the result to a Pandas DataFrame
london4s_df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
print(len(london4s_df))

# Display the first 10 rows of the DataFrame
london4s_df.head(10)
```
